```python
clientes_nuevos['timestamp'] = pd.to_datetime(clientes_nuevos['timestamp'])
```

```python
clientes_nuevos['dia'] = clientes_nuevos['timestamp'].dt.day
```

```python
clientes_nuevos['mes'] = clientes_nuevos['timestamp'].dt.month
```

```python
mes_minimo = clientes_nuevos.groupby('person')['mes'].min().to_frame()
```

```python
mes_minimo.reset_index()
```

```python
mes_minimo = mes_minimo.merge(clientes_nuevos, how = 'left', on = ['person', 'mes'])
```

```python
mes_minimo = mes_minimo.iloc[:,[0,2,3,4]]
```

```python
dia_minimo = mes_minimo.groupby('person')['dia'].min().to_frame()
```

```python
dia_minimo.reset_index()
```

```python
dia_minimo = dia_minimo.merge(mes_minimo, how = 'left', on = ['person', 'dia'])
```

```python
dia_minimo = dia_minimo.iloc[:,[0,2,3]]
```

```python
comienzo = dia_minimo.groupby('person')['timestamp'].min().to_frame()
```

```python
comienzo.rename(columns = {"timestamp" : "inicio_sesion"}).reset_index()
```

```python
fin = dia_minimo.groupby('person')['timestamp'].max().to_frame()
```

```python
fin.rename(columns = {"timestamp" : "fin_sesion"}).reset_index()
```

```python
duracion_sesion = comienzo.merge(fin, how = 'left', on = 'person')
```

```python
duracion_sesion.reset_index()
```

```python
duracion_sesion['duracion'] = duracion_sesion['timestamp_y'] - duracion_sesion['timestamp_x']
```

```python
duracion_sesion = duracion_sesion.iloc[:,2].to_frame()
```

```python
mes_minimo['flag_compra'] = np.where(mes_minimo['event'] == 'conversion', '1','0')
```

```python
flag_compra = mes_minimo.groupby('person')['flag_compra'].max().to_frame()
```

```python
flag_compra.reset_index()
```

```python
conversion_duracion = duracion_sesion.merge(flag_compra, how = 'left', on = 'person')
```

```python
conversion_duracion = conversion_duracion.reset_index()
```

```python
conversion_duracion.loc[conversion_duracion['flag_compra'] == '1']['duracion'].mean()
```
